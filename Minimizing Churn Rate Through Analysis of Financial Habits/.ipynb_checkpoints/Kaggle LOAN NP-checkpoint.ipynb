```python
# FEATURE ENGINEERING 
X['RATIO'] = X['AMT_CREDIT'] / X['AMT_GOODS_PRICE']
X_testkg['RATIO'] = X_testkg['AMT_CREDIT'] / X_testkg['AMT_GOODS_PRICE']

X['CREDIT_INCOME_PERCENT'] = X['AMT_CREDIT'] / X['AMT_INCOME_TOTAL']
X['ANNUITY_INCOME_PERCENT'] = X['AMT_ANNUITY'] / X['AMT_INCOME_TOTAL']
X['CREDIT_TERM'] = X['AMT_ANNUITY'] / X['AMT_CREDIT']
X['DAYS_EMPLOYED_PERCENT'] = X['DAYS_EMPLOYED'] / X['DAYS_BIRTH']

X_testkg['CREDIT_INCOME_PERCENT'] = X_testkg['AMT_CREDIT'] / X_testkg['AMT_INCOME_TOTAL']
X_testkg['ANNUITY_INCOME_PERCENT'] = X_testkg['AMT_ANNUITY'] / X_testkg['AMT_INCOME_TOTAL']
X_testkg['CREDIT_TERM'] = X_testkg['AMT_ANNUITY'] / X_testkg['AMT_CREDIT']
X_testkg['DAYS_EMPLOYED_PERCENT'] = X_testkg['DAYS_EMPLOYED'] / X_testkg['DAYS_BIRTH']
```

```python
#COLSCAN
def colscan(s):
    s = pd.Series(s)
    if len(s.unique())<100 :
        plt.figure(figsize=(20, 6))
        pd.Series(s).value_counts().plot('bar')
        plt.title(s.name)
        plt.show()
    else :
        print("Valeurs uniques >100 pas d'histogramme")
    tab= s.value_counts()
    tab = pd.DataFrame(tab)
    l = len(s)
    name = s.name
    print("----------- Liste des Valeurs -------")
    tab['Pourcent'] = tab[name]/l*100
    print(tab)
    print("Unique values = ", len(s.unique()))
    print("Len =", l)
    print("NAN = ", s.isna().sum())
    print("NULL = ", s.isnull().sum())
    print('Non NA', s.count())
    print("Dtype count =", s.get_dtype_counts())
    print("Type =", s.dtype)
    print("------Describe -----\n", s.describe(), "\n------End Describe---")
    


colscan(y)
```

```python
#Encodage des colonnes avec plusieurs 
le = LabelEncoder()
le_count = 0


for col in X:
    if X[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(X[col].unique())) <= 2:
            # Train on the training data
            le.fit(X[col])
            # Transform both training and testing data
            X[col] = le.transform(X[col])
            X_testkg[col] = le.transform(X_testkg[col])
            print("Encodage de la colonne", col)
            # Keep track of how many columns were label encoded
            le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
# Create an anomalous flag column
X['DAYS_EMPLOYED_ANOM'] = (X["DAYS_EMPLOYED"] == 365243)

# Replace the anomalous values with nan
X['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

X_testkg['DAYS_EMPLOYED_ANOM'] = X_testkg["DAYS_EMPLOYED"] == 365243
X_testkg["DAYS_EMPLOYED"].replace({365243: np.nan}, inplace = True)

print('There are %d anomalies in the train out of %d entries' % (X["DAYS_EMPLOYED_ANOM"].sum(), len(X)))
print('There are %d anomalies in the test data out of %d entries' % (X_testkg["DAYS_EMPLOYED_ANOM"].sum(), len(X_testkg)))

```

```python
from sklearn.preprocessing import MinMaxScaler, Imputer

X_train0 = X
    
# Feature names
features = list(X_train0.columns)

# Copy of the testing data
X_test0 = X_testkg

# Median imputation of missing values
imputer = Imputer(strategy = 'median')

# Scale each feature to 0-1
scaler = MinMaxScaler(feature_range = (0, 1))

# Fit on the training data
imputer.fit(X_train0)

# Transform both training and testing data
X_train0 = imputer.transform(X_train0)
X_test0 = imputer.transform(X_test0)

# Repeat with the scaler
scaler.fit(X_train0)
X_train0 = scaler.transform(X_train0)
X_test0 = scaler.transform(X_test0)

X_train0 = pd.DataFrame(X_train0, columns = X.columns)
X_test0 = pd.DataFrame(X_test0, columns = X_testkg.columns)

# drop des  SK ID CURR et stockage  part
skcoltrain = X_train0['SK_ID_CURR']
skcoltest = X_test0['SK_ID_CURR']
X_train0= X_train0.drop(columns=['SK_ID_CURR'])
X_test0 = X_test0.drop(columns=['SK_ID_CURR'])

print('Training data shape: ', X_train0.shape)
print('Target data shape: ', y.shape)
print('Testing data shape: ', X_test0.shape)
```
